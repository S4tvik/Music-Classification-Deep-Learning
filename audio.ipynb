```python

#getting training and testing dataset from the dataset 
tf.random.set_seed(1234)
train = tf.keras.preprocessing.image_dataset_from_directory(directory='Data/images_original/',validation_split=0.3,shuffle=True,subset="training",color_mode='rgb',seed=123,image_size=(200,200),batch_size=32,)
valid = tf.keras.preprocessing.image_dataset_from_directory(directory="Data/images_original/",validation_split=0.3,shuffle=True,subset="validation",color_mode="rgb",seed = 123, image_size=(200,200),batch_size=32)
train1 = tf.keras.preprocessing.image_dataset_from_directory(directory='Data/images_original/',validation_split=0.3,shuffle=True,subset="training",color_mode='rgb',seed=123,image_size=(400,400),batch_size=32,)
valid1 = tf.keras.preprocessing.image_dataset_from_directory(directory="Data/images_original/",validation_split=0.3,shuffle=True,subset="validation",color_mode="rgb",seed = 123, image_size=(400,400),batch_size=32)
```

```python
train.class_names
sample = train.take(1)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64,(3,3),activation='relu',input_shape = (200,200,3)),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(128,(3,3),activation='relu',),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=100, activation='relu',),

    tf.keras.layers.Dense(units=10,activation='softmax')
]
)
```

```python
model.compile(loss = 'sparse_categorical_crossentropy',optimizer='Adam',metrics=["accuracy"])
model.summary()
history = model.fit(train,validation_data = valid,epochs=10)
```

```python
model1 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64,(3,3),activation='relu',input_shape = (400,400,3)),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Conv2D(64,(3,3),activation='relu',),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=50,activation = 'relu'),
    tf.keras.layers.Dense(units=10,activation='softmax'),
]
)
```

```python
model1.compile(loss = 'sparse_categorical_crossentropy',optimizer='Adam',metrics=["accuracy"])
model1.summary()
cb = tf.keras.callbacks.EarlyStopping(monitor="val_loss",patience=3)
```

```python
history1 = model1.fit(train1,validation_data=valid1, epochs=10,batch_size=32,callbacks=cb)
```

```python
plt.plot(history1.history['accuracy'])

plt.plot(history1.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
