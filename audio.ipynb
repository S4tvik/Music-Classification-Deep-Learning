```python

#getting training and testing dataset from the dataset 
tf.random.set_seed(1234)
train = tf.keras.preprocessing.image_dataset_from_directory(directory='Data/images_original/',validation_split=0.3,shuffle=True,subset="training",color_mode='rgb',seed=123,image_size=(256,256),batch_size=32,)
valid = tf.keras.preprocessing.image_dataset_from_directory(directory="Data/images_original/",validation_split=0.3,shuffle=True,subset="validation",color_mode="rgb",seed = 123, image_size=(256,256),batch_size=32)
train1 = tf.keras.preprocessing.image_dataset_from_directory(directory='Data/images_original/',validation_split=0.3,shuffle=True,subset="training",color_mode='rgb',seed=123,image_size=(400,400),batch_size=32,)
valid1 = tf.keras.preprocessing.image_dataset_from_directory(directory="Data/images_original/",validation_split=0.3,shuffle=True,subset="validation",color_mode="rgb",seed = 123, image_size=(400,400),batch_size=32)
```

```python
train.class_names
train 
sample = train.take(1)
sample
import matplotlib.pyplot as plt

# Assuming the dataset is using RGB images
for images, labels in sample:
    # Visualize the first few images in the batch
    num_images_to_visualize = min(images.shape[0], 5)
    
    for i in range(num_images_to_visualize):
        plt.subplot(1, num_images_to_visualize, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))  # Convert to uint8 for Matplotlib
        plt.title(f"Label: {labels[i].numpy()}")
        plt.axis("off")

    plt.show()
```

Creating a 5 layer CNN model (256, 128, 64, 64, 32)
Dropout and max pooling after each layer to avoid overfitting of the model.

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(256,(3,3),activation='relu',input_shape = (256,256,3)),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Conv2D(128,(3,3),activation='relu',),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Conv2D(64,(3,3),activation='relu',),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Conv2D(64,(3,3),activation='relu',),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Conv2D(32,(3,3),activation='relu',),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128, activation='relu',),
    tf.keras.layers.Dense(units=10,activation='softmax')
]
)
```

```python
model.compile(loss = 'sparse_categorical_crossentropy',optimizer='Adam',metrics=["accuracy"])
model.summary()
history = model.fit(train,validation_data = valid,epochs=25)
```

```python
plt.plot(history.history['accuracy'])

plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
